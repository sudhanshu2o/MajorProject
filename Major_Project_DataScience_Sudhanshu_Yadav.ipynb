```python
#NORMALIZATION of Values of Mileage
from sklearn.preprocessing import MinMaxScaler
data_Mileage=data["Mileage"].values.reshape(-1,1)
scale=MinMaxScaler()
scale=scale.fit(data_Mileage)
data_Mileage_norm=scale.transform(data_Mileage)
data["Mileage"]=data_Mileage_norm
data.head()
```
